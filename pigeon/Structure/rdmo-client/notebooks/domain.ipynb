```python
attributes = client.list_attributes()
```

```python
def build_tree(attributes):
    # Create a lookup dictionary by ID
    attr_dict = {attr['id']: {**attr, 'children': []} for attr in attributes}
    root_nodes = []

    for attr in attr_dict.values():
        parent_id = attr.get('parent')
        if parent_id and parent_id in attr_dict:
            # Add this node to its parent's children list
            attr_dict[parent_id]['children'].append(attr)
        else:
            # It's a top-level node
            root_nodes.append(attr)
    return root_nodes

# Usage:
tree = build_tree(attributes)
```
